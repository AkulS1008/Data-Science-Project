```python
import pandas as pd

car_units = pd.DataFrame({"Model" : ["BMV", "Honda", "Toyota", "Volkswagen", "Infinity", "Porshe", "Car_MODEL_A"], "Units" : [1, 2, 4, 3, 4, 5, 6]})
car_costs = pd.DataFrame({"Model" : ["BMV", "Honda", "Toyota", "Volkswagen", "Infinity", "Porshe", "Acura"], "Unit Cost" : [10000, 20000, 45000, 900000, 30000, 90000, 20000]})
```

```python
pd.merge(car_units, car_costs, how = 'right', indicator = True) #Takes intersection of columns 
```

```python
car_models = pd.DataFrame({"Unit" : [1, 2, 3, 4, 5, 6, 7], "Model" : ["BMV", "Honda", "Toyota", "Volkswagen", "Infinity", "Porshe", "Car_MODEL_A"]})
car_units = pd.DataFrame({"Unit" : [1, 2, 3, 4, 5, 6, 7], "Model" : ["BMV", "Honda", "Toyota", "Volkswagen", "Infinity", "Porshe", "Car_MODEL_A"]})
car_costs = pd.DataFrame({"Unit" : [1, 2, 3, 4, 5, 6, 7], "Model" : ["BMV", "Honda", "Toyota", "Volkswagen", "Infinity", "Porshe", "Car_MODEL_A"]})
res = pd.concat([car_models, car_units, car_costs])#.drop_duplicates() #Looks for index in sequence and same column names
res1 = res.groupby("Model")
keys = res1.groups.keys()
for i in keys:
    print(res1.get_group(i))
```
