```python
import numpy as np
import pandas as pd

result = pd.read_csv('WHO-COVID-19-global-data.csv')
result1 = result.loc[(result.New_cases > 0) & (result.Cumulative_cases > 0)]
result1.reset_index(inplace = True)
result1['per_increase'] = (result1.New_cases/result1.Cumulative_cases) * 100
result1 = result1.sort_values(by = "per_increase")
result1
#result.loc[result.Country_code == 'US', ['New_cases']].max() #Max new cases in US without idxmax
```
